```
import pandas as pd
df = pd.read_csv(url)

df_not_null = df[df.year.notnull()]
df_not_null
```

```
import pandas as pd
df = pd.read_csv(url)

years = [2002,2008]
selection = df[df.year.isin(years)]

selection
```

Select Row based not in the list

```
import pandas as pd
df = pd.read_csv(url)

years = [2002,2008]
selection = df[~df.year.isin(years)]

selection
```

Select data between 2004 and 2008 that are from singapore

```
import pandas as pd
df = pd.read_csv(url)

years = [2002,2008]
country = ['Singapore']
selection = df[df.year.isin(years) & df.country.isin(country)]

selection
```

Select data between 2004 and 2008 that are not from singapore

```
import pandas as pd
df = pd.read_csv(url)

years = [2002,2008]
country = ['Singapore']
selection = df[df.year.isin(years) & ~df.country.isin(country)]

selection
```
